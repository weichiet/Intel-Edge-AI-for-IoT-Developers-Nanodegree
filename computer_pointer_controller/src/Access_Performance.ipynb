# Computer Pointer Controller

## Access Performance

```python
import matplotlib.pyplot as plt

device = 'CPU'
precision_list = ["FP32", "FP16", "INT8"]
inference_time=[]
fps=[]
model_load_time=[]

for precision in precision_list:
    with open('../bin/CPU/'+ precision + '/stats.txt', 'r') as f:
        inference_time.append(float(f.readline().split("\n")[0]))
        fps.append(float(f.readline().split("\n")[0]))
        model_load_time.append(float(f.readline().split("\n")[0]))
```

```python
plt.bar(precision_list, inference_time)
plt.xlabel("Precision")
plt.ylabel("Time in Seconds \n (The Lower the Better)")
plt.title("Total Inference Time - CPU")
plt.show()
```

```python
plt.bar(precision_list, fps)
plt.xlabel("Precision")
plt.ylabel("FPS\n (The Higher the Better)")
plt.title("Frames per Second - CPU")
plt.show()
```

```python
plt.bar(precision_list, model_load_time)
plt.xlabel("Precision")
plt.ylabel("Time in Seconds \n (The Lower the Better)")
plt.title("Model Loading Time - CPU")
plt.show()
```
